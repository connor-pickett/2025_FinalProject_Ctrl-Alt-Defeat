WR Lists with Calculated Variables

```python
season_2019= nfl.import_seasonal_pfr("rec",[2019]) #get 2019 data
WRs_2019 = season_2019[season_2019["pos"] == "WR"] #sort by wide recievers
WRs_2019 = WRs_2019[WRs_2019["tm"] != "2TM"]  
WRs_2019 = WRs_2019[WRs_2019["tm"] != "3TM"] # exlude all players who got traded
WRs_2019 = WRs_2019[WRs_2019["tm"] != "4TM"] 
WRs_2019["rec_share"] = WRs_2019.groupby("tm")['rec'].transform(lambda x:x/x.sum()*100)
WRs_2019["tgt_share"] = WRs_2019.groupby("tm")['tgt'].transform(lambda x:x/x.sum()*100) #calculate our key variables
WRs_2019["td_share"] = WRs_2019.groupby("tm")['td'].transform(lambda x:x/x.sum()*100)
WRs_2019["yds_share"] = WRs_2019.groupby("tm")['yds'].transform(lambda x:x/x.sum()*100)
```

Make WR and QB list and their horoscopes

```python
def get_player_list(position):
    rost = nfl.import_seasonal_rosters([2019, 2020, 2021, 2022, 2023, 2024])
    filtered = rost[rost["position"] == position]
    return filtered[["season","player_name", "position", "team", "birth_date"]]
```

```python
get_player_list("WR")
get_player_list("QB")
```

```python
# Function to get zodiac sign
def get_zodiac_sign(birthdate):
    if pd.isnull(birthdate):
        return None
    birthdate = pd.to_datetime(birthdate)
    month = birthdate.month
    day = birthdate.day

    if (month == 1 and day >= 20) or (month == 2 and day <= 18):
        return "Aquarius"
    elif (month == 2 and day >= 19) or (month == 3 and day <= 20):
        return "Pisces"
    elif (month == 3 and day >= 21) or (month == 4 and day <= 19):
        return "Aries"
    elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
        return "Taurus"
    elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
        return "Gemini"
    elif (month == 6 and day >= 21) or (month == 7 and day <= 22):
        return "Cancer"
    elif (month == 7 and day >= 23) or (month == 8 and day <= 22):
        return "Leo"
    elif (month == 8 and day >= 23) or (month == 9 and day <= 22):
        return "Virgo"
    elif (month == 9 and day >= 23) or (month == 10 and day <= 22):
        return "Libra"
    elif (month == 10 and day >= 23) or (month == 11 and day <= 21):
        return "Scorpio"
    elif (month == 11 and day >= 22) or (month == 12 and day <= 21):
        return "Sagittarius"
    elif (month == 12 and day >= 22) or (month == 1 and day <= 19):
        return "Capricorn"

# Function to get player zodiac signs
def get_player_zodiac_data(position):
    roster = nfl.import_seasonal_rosters([2019, 2020, 2021, 2022, 2023, 2024])

    # Filter the data to only include the specified position
    position_data = roster[roster["position"] == position]

    player_data = position_data[["player_name", "position", "team", "birth_date", "season"]]

    # Drop rows without birth dates
    player_data = player_data.dropna(subset=["birth_date"])

    # Add zodiac sign column
    player_data["zodiac_sign"] = player_data["birth_date"].apply(get_zodiac_sign)

    return player_data

# Run it
wr_df = get_player_zodiac_data("WR")
```

```python
def get_means(df, sign_x, sign_y):
    filtered_df = df[
        (df["zodiac_sign_x"] == sign_x) & 
        (df["zodiac_sign_y"] == sign_y)
    ]
    numb_instances = len(filtered_df)
    zodiac_combo = f"{sign_x}-{sign_y}"
    WR_sign = sign_x
    QB_sign = sign_y

    # Means
    rec_share_mean = filtered_df["rec_share"].mean()
    td_share_mean = filtered_df["td_share"].mean()
    tgt_share_mean = filtered_df["tgt_share"].mean()
    yds_share_mean = filtered_df["yds_share"].mean()

    # Standard Deviations
    rec_share_std = filtered_df["rec_share"].std()
    td_share_std = filtered_df["td_share"].std()
    tgt_share_std = filtered_df["tgt_share"].std()
    yds_share_std = filtered_df["yds_share"].std()

    #Simulated
    import numpy as np
    simulated_data_rec = np.random.normal(loc= rec_share_mean, scale= rec_share_std, size=100000)
    rec_share_sim_mean= simulated_data_rec.mean()
    simulated_data_td = np.random.normal(loc= td_share_mean, scale= td_share_std, size=100000)
    td_share_sim_mean= simulated_data_td.mean()
    simulated_data_tgt = np.random.normal(loc= tgt_share_mean, scale= tgt_share_std, size=100000)
    tgt_share_sim_mean= simulated_data_tgt.mean()
    simulated_data_yds = np.random.normal(loc= yds_share_mean, scale= yds_share_std, size=100000)
    yds_share_sim_mean= simulated_data_yds.mean()

    return pd.DataFrame({
        "zodiac_combo": [zodiac_combo],
        "wr_sign": [WR_sign],
        "qb_sign": [QB_sign],
        "instances_number": [numb_instances],
        "rec_share_mean": [rec_share_mean],
        "rec_share_std": [rec_share_std],
        "rec_share_simulated_mean": [rec_share_sim_mean],
        "td_share_mean": [td_share_mean],
        "td_share_std": [td_share_std],
        "td_share_simulated_mean": [td_share_sim_mean],
        "tgt_share_mean": [tgt_share_mean],
        "tgt_share_std": [tgt_share_std],
        "tgt_share_simulated_mean": [tgt_share_sim_mean],
        "yds_share_mean": [yds_share_mean],
        "yds_share_std": [yds_share_std],
        "yds_share_simulated_mean": [yds_share_sim_mean]
 
    })
```

```python
import numpy as np
import pandas as pd

def custom_t_test(pair_mean, zodiac_combo, mean=15.349357, std=6.664931, sims=100):
    np.random.seed(42)
    simulated_data = np.random.normal(loc=mean, scale=std, size=sims)

    # Calculate t-statistic using standard error
    standard_error = simulated_data.std(ddof=1) / np.sqrt(sims)
    t_stat = (simulated_data.mean() - pair_mean) / standard_error

    # Calculate p-value: proportion of simulations less than the pair_mean
    p_value = np.mean(simulated_data < pair_mean)

    # Return result as a DataFrame row
    return pd.DataFrame({
        "Zodiac Combo": [zodiac_combo],
        "Pair Mean": [pair_mean],
        "Simulated Mean": [simulated_data.mean()],
        "T-Stat": [t_stat],
        "P-Value": [p_value]
    })
```
